```python
import csv
import requests as rq
import pandas as pd
from pandas import DataFrame, Series
import re
import csv
import json
import pylab
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
columns = ['MSN', 'StateCode', 'Year', 'Data']
csvfile = pd.read_csv('ProblemCData.csv',encoding = 'utf-8',header=0,names=columns)
```

```python
all_variable = {}
```

```python
for idx, row in csvfile.iterrows():
    if not row.MSN in all_variable:
        new_variable = row.MSN
        all_variable[new_variable] = {}
        all_variable[new_variable][row.StateCode] = {}
        all_variable[new_variable][row.StateCode][row.Year] = row.Data
    else:
        if not row.StateCode in all_variable[row.MSN]:
            new_state = row.StateCode
            all_variable[row.MSN][new_state] = {}
            all_variable[row.MSN][new_state][row.Year] = row.Data
        else:
            all_variable[row.MSN][row.StateCode][row.Year] = row.Data
```

```python
years = range(1960,2010)
plt.figure()
for state in all_variable['ABICB']:
    plt.plot(years,list(all_variable['ABICB'][state].values()))
```
